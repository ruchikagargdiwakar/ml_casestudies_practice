# Read the data file and check for any missing values

```python
import pandas as pd

df = pd.read_csv("Project_Data_2.csv")
df.head()
```

# Change the headers to country and year accordingly.

```python
df.columns = ['Country','Wheat sales in 1990(tons)','Wheat sales in 1991(tons)','Wheat sales in 1992(tons)',
              'Wheat sales in 1993(tons)','Wheat sales in 1994(tons)','Wheat sales in 1995(tons)','Wheat sales in 1996(tons)'
             ,'Wheat sales in 1997(tons)','Wheat sales in 1998(tons)','Wheat sales in 1999(tons)','Wheat sales in 2000(tons)'
             ,'Wheat sales in 2001(tons)','Wheat sales in 2002(tons)','Wheat sales in 2003(tons)','Wheat sales in 2004(tons)'
             ,'Wheat sales in 2005(tons)','Wheat sales in 2006(tons)','Wheat sales in 2007(tons)']
df.head()
```

# Cleanse the data if required and remove null or blank values

```python
#df.iloc[:,1:]
```

```python
#print(df.isna().sum())
df=df.fillna(0)
#df.head()
#print(df.isna().sum())
df.iloc[:,1:] = df.iloc[:,1:].stack().astype(str).str.replace(',','.', regex=True).unstack().astype(float)
df.head()
```

# After the EDA part is done, try to think which algorithm should be applied here.
As we need to make this across years we need to apply PCA first.

Apply PCA on the dataset and find the number of principal components which explain nearly all the variance.

```python
from sklearn.decomposition import PCA
import pandas as pd

pca = PCA(n_components = 6)

pca.fit(df.iloc[:,1:])

#reduced features data
reduced_features_data = pca.transform(df.iloc[:,1:])
reduced_features_df = pd.DataFrame(reduced_features_data)

#reduced_features_df.columns = ['PC1','PC2','PC3''PC4''PC4']

print(reduced_features_df.head())

#variance ratio
print(pca.explained_variance_ratio_)
```

# Plot elbow chart or scree plot to find out optimal number of clusters.

```python
from sklearn import metrics
from scipy.spatial.distance import cdist
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

distortions = []
K = range (1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(reduced_features_df)
    kmeanModel.fit(reduced_features_df)
    distortions.append(sum(np.min(cdist(reduced_features_df,kmeanModel.cluster_centers_,'euclidean'),axis=1))/reduced_features_df.shape[0])
    
plt.plot(K,distortions,'bx-')
plt.xlabel('k')
plt.ylabel('Distortions')
plt.title('The elbow method showing the optimal k')
plt.show()
```

# Then try to apply K means, Hierarchical clustering and showcase the results.

```python
kmeanModel = KMeans(n_clusters=2)
#kmeanModel.fit(reduced_features_df)
kmeanModel.fit_predict(reduced_features_df)

# cluster center vectors
print(kmeanModel.cluster_centers_)
print(kmeanModel.labels_)
print(len(kmeanModel.labels_))
#df['clusters'] = kmeanModel.labels_
#print(df['clusters'])
```

```python
from sklearn.cluster import AgglomerativeClustering
import numpy as np
X = reduced_features_df.values
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(X)
print(cluster.labels_)
```
