# The data file contains numerical attributes that describe a letter and its corresponding class. Read the datafile “letterCG.data” and set all the numerical attributes as features.Split the data in to train and test sets.


```python
import numpy as np
import pandas as pd

df = pd.read_csv("letterCG.bin",delimiter=" ")
df.head()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df['Class'])
le.classes_
df['Class'] = le.transform(df['Class']) 
print(df.head(5))

#check for co-relation
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar=True, annot=True, square = True, cmap = 'coolwarm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df.iloc[:,1:16], df['Class'], test_size=0.20, random_state=5)
```

# Fit a sequence of AdaBoostClassifier with varying number of weak learners ranging from 1 to 16, keeping the max_depth as 1. Use decision tree classifier as the base classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier

from sklearn import metrics

classifiers = [AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1), 
                n_estimators=16,random_state=7)]

for i in range(1, 17):
    print("="*30)
    print("range::",i)
    clf = AdaBoostClassifier(base_estimator=DecisionTreeClassifier(max_depth=1),n_estimators=i,random_state=7)
    clf.fit(train_x,train_y)
    pred_y = clf.predict(test_x)
    #calculate accuracy
    calc_accuracy = metrics.accuracy_score(test_y,pred_y)
    print("Accuracy of model::", calc_accuracy)
```
