# John Cancer Hospital (JCH) is a leading cancer hospital in USA. It specializes in preventingbreast cancer. Over the period of the last few years, JCH has collected breast cancer data from patients who came for screening/treatment.However,this data has almost 30 attributes and is difficult to run and interpret the result. You as ML expert has to reduce the no. of attributes (Dimensionality Reduction) so that results are meaningful and accurate

```python
import pandas as pd

df = pd.read_csv("breast-cancer-data.csv")
df.head(5)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df['diagnosis'])
le.classes_
df['diagnosis'] = le.transform(df['diagnosis']) 
print(df.head(5))
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df, df['diagnosis'], test_size=0.20, random_state=5)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
lda.fit(train_x, train_y)
train_x = lda.transform(train_x)
test_x = lda.transform(test_x)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC,NuSVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier

from sklearn import metrics

classifiers = [KNeighborsClassifier(3), DecisionTreeClassifier(), RandomForestClassifier(), 
               SVC(kernel='rbf',C=0.025,probability=True),
               NuSVC(probability= True),
               GaussianNB(),
               AdaBoostClassifier(n_estimators=100,random_state=7)]

for clf in classifiers:
    name = clf.__class__.__name__
    print("="*30)
    print(name)
    clf.fit(train_x,train_y)
    pred_y = clf.predict(test_x)
    #calculate accuracy
    calc_accuracy = metrics.accuracy_score(test_y,pred_y)
    print("Accuracy of model::", calc_accuracy)
```
