# PeerLoanKart is anNBFC (Non-BankingFinancial Company) which facilitates peer to peer loan. It connects people who need money (borrowers) with people who have money (investors). As an investor,you would want to invest in people who showed a profile of having a high probability of paying you back. You as a ML expert create a model that will help predict whether a borrower will pay the loan or not. 

```python
import pandas as pd

df = pd.read_csv("loan_borowwer_data.csv")
df.head(5)
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
le.fit(df['purpose'])
le.classes_
df['purpose'] = le.transform(df['purpose']) 
#print(df)

#check for co-relation
import matplotlib.pyplot as plt
import seaborn as sns
corr = df.corr()
plt.figure(figsize=(14,14))
sns.heatmap(corr, cbar=True, annot=True, square = True, cmap = 'coolwarm')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
features = ['purpose','int.rate','installment','log.annual.inc','dti','delinq.2yrs','pub.rec','not.fully.paid']
train_x, test_x, train_y, test_y = train_test_split(df,df['credit.policy'],test_size=0.20, random_state=5)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

classifiers = [DecisionTreeClassifier(), RandomForestClassifier()]

for clf in classifiers:
    name = clf.__class__.__name__
    print("="*30)
    print(name)
    clf.fit(train_x,train_y)
    pred_y = clf.predict(test_x)
    #calculate accuracy
    calc_accuracy = metrics.accuracy_score(test_y,pred_y)
    print("Accuracy of model::", calc_accuracy)
```

```python
#Apply LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
lda.fit(train_x,train_y)
transformed_train_x = lda.transform(train_x)
print("Number of components in transformed shape {}".format(transformed_train_x.shape[1]))

transformed_test_x = lda.transform(test_x)
print("Number of components in transformed shape {}".format(transformed_test_x.shape[1]))
```

```python
for clf in classifiers:
    name = clf.__class__.__name__
    print("="*30)
    print(name)
    clf.fit(transformed_train_x,train_y)
    pred_y = clf.predict(transformed_test_x)
    #calculate accuracy
    calc_accuracy = metrics.accuracy_score(test_y,pred_y)
    print("Accuracy of model::", calc_accuracy)
```

As per my understanding, models are working better without LDA transformation.
