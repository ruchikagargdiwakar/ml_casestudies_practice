# Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes. 

```python
df['total_benefitted'] = (df[df.columns[2:6]]).sum(axis=1)
df_copy = df.copy()
df
```

# Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states. 

```python
df.loc['Total'] = pd.Series(df['total_benefitted'].sum(), index = ['total_benefitted'])
print (df)
```

# Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

```python
label = df_copy['STATE/UT']
benefitted_inmates = df_copy['total_benefitted']

def plot_bar_x():
    index = np.arange(len(label))
    plt.bar(index, benefitted_inmates)
    plt.xlabel('States', fontsize=5)
    plt.ylabel('Total Benefitted inmates', fontsize=5)
    plt.xticks(index, label, fontsize=5, rotation=90)
    plt.title('Benefitted Inmates for each states')
    plt.show()
```

```python
plot_bar_x()
```

# Make a pie chart that depicts the ratio among different modes of benefits.

```python
labels = df.columns[2:6]
#print(labels)

sizes = [df['No. of Inmates benefitted by Elementary Education'].sum(),df['No. of Inmates benefitted by Adult Education'].sum(),df['No. of Inmates benefitted by Higher Education'].sum(),df['No. of Inmates benefitted by Computer Course'].sum()]
#print(sizes)

fig1, ax1 = plt.subplots()
explode = (0, 0, 0, 0)
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
